```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("PreProcessedTargetVolatilityHuobi2020-03-01_1000111111111111111.csv")
dataset.head()
```

```python
dataset["IsWeekend"].dtypes
```

```python
dataset.nunique()
```

```python
dataset.IsWeekend.unique()
```

```python
dataset["IsWeekend"].replace(to_replace=4.1245,value=1, inplace=True)
```

```python
dataset.head()
```

```python
dataset=dataset.drop(["DateTime", "Hour"],axis=1)
dataset.head()
```

```python
train=dataset.iloc[:,:-1]
train.head()
```

```python
y=dataset["VolatilityTargetFuture5Minutes"]
```

```python
import statsmodels.api as sm
```

```python
results=sm.OLS(y,train).fit()
```

```python
results.summary()
```

```python
results.params
```

```python
newTrain=train.drop(["ImpactSpreadDot5Deep","ImpactSpreadDot2Deep"],axis=1)
newResults=sm.OLS(y,newTrain).fit()
newResults.summary()
```

```python
newResults.params
```

```python
dropOneDeepTrain=newTrain.drop(["ImpactSpread1Deep"],axis=1)
dropOneResults=sm.OLS(y,dropOneDeepTrain).fit()
dropOneResults.summary()
```

```python
dropOneResults.params
```

```python
dropAskBid=dropOneDeepTrain.drop(["AskBidVolume"], axis=1)
dropBidAndAskResults=sm.OLS(y,dropAskBid).fit()
dropBidAndAskResults.summary()
```

```python
dropBidAndAskResults.params
```

```python
dropISWeekend=dropAskBid.drop(["IsWeekend"], axis=1)
dropWeekendResults=sm.OLS(y,dropISWeekend).fit()
dropWeekendResults.summary()
```

```python
dropWeekendResults.params
```

```python
dropvolTen=dropISWeekend.drop(["Volatility10Seconds"], axis=1)
dropTenRes=sm.OLS(y,dropvolTen).fit()
dropTenRes.summary()
```

```python
dropTenRes.params
```

```python
dropDot1Deep=dropvolTen.drop(["ImpactSpreadDot1Deep"], axis=1)
dropDotOneRes=sm.OLS(y,dropDot1Deep).fit()
dropDotOneRes.summary()
```

```python
dropDotOneRes.params
```

```python
dropVol5sec=dropDot1Deep.drop(["Volatility5Seconds"], axis=1)
dropDotOneRes=sm.OLS(y,dropVol5sec).fit()
dropDotOneRes.summary()
```
